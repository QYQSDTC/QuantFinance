```python
# 计算ths_data的3日，5日和10日收益率
ths_data['3日收益率'] = ths_data['close'].pct_change(3) * 100
ths_data['5日收益率'] = ths_data['close'].pct_change(5) * 100
ths_data['10日收益率'] = ths_data['close'].pct_change(10) * 100

ths_data.tail()
```

由于qstock无法获取同花顺概念板块的所有成份股，故无法根据概念板块成份股数目进行筛选，这里转换思路直接计算所有概念板块的3，5，和10日平均收益率然后进行排序，各取其TOP20.

```python
'''
Author: Yiqian Qian
Description: file content
Date: 2022-10-13 00:41:24
LastEditors: Yiqian Qian
LastEditTime: 2024-03-06 20:21:15
FilePath: /lesson2/同花顺/同花顺行业概念EDA.ipynb
'''
return_dict = {}
cnt = 0
suc = 0
max_retries = 2  # 最大重试次数
retry_delay = 300  # 重试延迟（秒）

for index in index_list:
    attempts = 0
    while attempts < max_retries:
        try:
            data = qs.ths_index_data(index)
            # 计算3日，5日和10日收益率
            data['return_3'] = data['close'].pct_change(3) * 100
            data['return_5'] = data['close'].pct_change(5) * 100
            data['return_10'] = data['close'].pct_change(10) * 100
            # 把计算好的收益率取最后一天存入字典
            return_dict[index] = data.iloc[-1, -4:]
            suc += 1
            break  # 成功后跳出循环
        except Exception as e:
            attempts += 1
            cnt += 1
            print(f'Error with {index}: {e}')
            if attempts < max_retries:
                print(f'Retrying {index}, attempt {attempts} of {max_retries}')
                time.sleep(retry_delay)
            else:
                print(f'{index} failed after {max_retries} attempts')
    print(f'{suc/len(index_list) * 100:.2f}% has been done')


return_dict
```

```python
ths_df = pd.DataFrame(return_dict).T
ths_df
```

# 热门股池
从同花顺概念板块中选出成交量前5的板块，再从这 5 个板块中选出成交量前 5 的标的

```python
# select top 5 concepts with highest volume
top10 = ths_df.sort_values('volume', ascending=False).head(10).index.values
concepts_exclude = ['沪股通', '标普道琼斯A股', '深股通'] # exclude these concepts
top5 = [x for x in top10 if x not in concepts_exclude][:5]
```

```python
top5
```

```python
# get the top 5 members with highest volume in these 5 concepts
hot_member = pd.DataFrame()
for concept in top5:
    memeber = qs.ths_index_member(concept)
    # sort with volume
    memeber5 = memeber.sort_values('成交额(亿)', ascending=False).head(5)
    hot_member = pd.concat([hot_member, memeber5])
hot_member
```

```python
# sort with volume
hot_member = hot_member.sort_values('成交额(亿)', ascending=False)
# reset index and drop old index
hot_member = hot_member.reset_index(drop=True)
# drop dupicate according to 代码 column
hot_member = hot_member.drop_duplicates(subset='代码')
hot_member
```

```python
fig = px.bar(hot_member, x='名称', y='成交额(亿)', color='涨跌幅', color_continuous_scale='oryel', range_color=(0, 20))
fig
```

```python
fig.write_html('hot_member.html')
fig.write_image('hot_member.png', scale=5)
```
